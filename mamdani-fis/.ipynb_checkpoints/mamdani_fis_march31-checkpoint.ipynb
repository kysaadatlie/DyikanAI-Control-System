## 2. FLC Architecture: Inputs, Outputs, and Design Rationale

The Mamdani FIS controls **three actuator groups**. Each group has its own set of relevant inputs:

| Actuator | FLC Inputs | Output |
|---|---|---|
| Heater / Fan | `air_temp_c`, `outdoor_temp_c`, `frost_risk` | Heater intensity (0–100) |
| Water pump | `soil_moisture_pct`, `air_temp_c`, `vpd_hpa` | Pump duration (0–120 s) |
| Grow lamp | `light_raw`, `air_temp_c` | Lamp power (0–100) |

**Why Mamdani over Sugeno?**
Mamdani uses fuzzy output sets and defuzzifies with centroid, producing smooth, continuously varying outputs — appropriate for physical actuators where abrupt switching causes wear. Sugeno uses crisp weighted sums, which is computationally cheaper but loses the interpretability that is essential for justifying agronomic rules in a diploma context.

**Why these inputs?**
- `air_temp_c`: primary control variable for climate management
- `outdoor_temp_c`: from Alfiia's predictions — heating effort needed depends heavily on external conditions in Naryn (mean outdoor temp = 1.9°C, min = −31.9°C)
- `frost_risk`: binary signal from Alfiia's ML model — triggers emergency heating
- `soil_moisture_pct`: direct irrigation trigger (0–68% range in dataset)
- `vpd_hpa`: vapor pressure deficit — high VPD means plants transpire fast and need more water
- `light_raw`: light sensor reading for lamp control

## 3. Universe of Discourse and Membership Function Definitions

Each input and output variable is mapped over a **universe of discourse** — the numerical range it can take. Membership functions (MFs) partition this range into overlapping linguistic terms. We use **triangular (trimf)** and **trapezoidal (trapmf)** MFs:
- Triangular: for interior terms where a single peak best represents the concept
- Trapezoidal: for boundary terms (COLD, HOT, DRY, WET) where the extreme values all belong equally to the set

```python
# ============================================================
#  UNIVERSES OF DISCOURSE
# ============================================================

# --- INPUT UNIVERSES ---
# Air temperature: extended range covering Naryn climate + greenhouse limits
# Sensor data: 21.6–23.7°C. Predicted data: -10.2–39.9°C
u_temp     = np.arange(-5, 46, 0.5)     # °C, greenhouse + outdoor range

# Outdoor temperature: Naryn Kyrgyzstan climate, harsh winters
# Predicted data: -31.9 to 34.4°C, mean=1.9°C
u_out_temp = np.arange(-35, 36, 0.5)    # °C

# Soil moisture: expressed as percentage (0–100%)
# Predicted data: 0–68.4%, mean=29.9%. We extend to 100 for completeness.
u_soil     = np.arange(0, 101, 1)       # %

# VPD (vapor pressure deficit): predicted data 0.7–57.6 hPa
# Normal greenhouse range: 5–25 hPa; above 25 = plant stress
u_vpd      = np.arange(0, 51, 0.5)      # hPa

# Light: raw sensor values from 3 (dark room at night) to ~2321 (bright day)
# Sensor data range 3–6 (nighttime readings), predicted: 2–2321
u_light    = np.arange(0, 2401, 10)     # raw lux units

# Frost risk: binary 0 or 1 from ML model
u_frost    = np.arange(0, 1.1, 0.1)

# --- OUTPUT UNIVERSES ---
# Heater control intensity (0 = off, 100 = full power)
# Defuzzified value is thresholded in Python: >40 → ON, >70 → ON + alert
u_heater   = np.arange(0, 101, 1)       # 0–100 intensity

# Pump duration in seconds (0 = no watering, 120 = full cycle)
# Your automation_engine uses 30s default; we extend to 120s for severe dryness
u_pump     = np.arange(0, 121, 1)       # seconds

# Lamp power (0 = off, 100 = full brightness)
u_lamp     = np.arange(0, 101, 1)

print('Universe of discourse summary:')
print(f'  Air temperature:    {u_temp.min():.0f} to {u_temp.max():.0f} °C    (step 0.5°C)')
print(f'  Outdoor temperature:{u_out_temp.min():.0f} to {u_out_temp.max():.0f} °C   (step 0.5°C)')
print(f'  Soil moisture:      {u_soil.min():.0f} to {u_soil.max():.0f} %    (step 1%)')
print(f'  VPD:                {u_vpd.min():.0f} to {u_vpd.max():.0f} hPa  (step 0.5)')
print(f'  Light:              {u_light.min():.0f} to {u_light.max():.0f} raw  (step 10)')
print(f'  Frost risk:         {u_frost.min():.0f} to {u_frost.max():.0f}       (binary via MF)')
print(f'  Heater output:      {u_heater.min():.0f} to {u_heater.max():.0f}       (0=off, 100=full)')
print(f'  Pump duration:      {u_pump.min():.0f} to {u_pump.max():.0f} sec  (step 1s)')
print(f'  Lamp output:        {u_lamp.min():.0f} to {u_lamp.max():.0f}       (0=off, 100=full)')
```

```python
# ============================================================
#  MEMBERSHIP FUNCTIONS — AIR TEMPERATURE (input)
# ============================================================
# Breakpoints justified by:
#   - Predicted data p5=12.8°C, p95=27.9°C
#   - Optimal greenhouse temp for common vegetables: 18–26°C
#   - Frost damage threshold: below 5°C
#   - Heat stress threshold: above 32°C

mf_temp = {
    'cold':  fuzz.trapmf(u_temp, [-5, -5, 8,  15]),   # frost risk zone
    'cool':  fuzz.trimf( u_temp, [10, 16, 22]),        # below optimal
    'ideal': fuzz.trimf( u_temp, [18, 23, 28]),        # target range
    'warm':  fuzz.trimf( u_temp, [24, 29, 35]),        # above optimal
    'hot':   fuzz.trapmf(u_temp, [31, 36, 45, 45]),   # heat stress
}

# ============================================================
#  MEMBERSHIP FUNCTIONS — OUTDOOR TEMPERATURE (input)
# ============================================================
# Breakpoints justified by:
#   - Naryn outdoor temp: mean=1.9°C, min=-31.9°C, max=34.4°C
#   - Heater effort scales with outdoor temp (bigger delta = more heat needed)

mf_out_temp = {
    'very_cold': fuzz.trapmf(u_out_temp, [-35, -35, -15, -5]),  # harsh Kyrgyz winter
    'cold':      fuzz.trimf( u_out_temp, [-10, -2,   8]),       # typical Naryn winter
    'mild':      fuzz.trimf( u_out_temp, [  3, 12,  22]),       # spring / autumn
    'warm':      fuzz.trapmf(u_out_temp, [ 18, 25,  35, 35]),   # summer outdoor
}

# ============================================================
#  MEMBERSHIP FUNCTIONS — SOIL MOISTURE (input)
# ============================================================
# Breakpoints justified by:
#   - Predicted data range: 0–68.4%, mean=29.9%
#   - Automation engine thresholds: needs_water >380 raw ≈ <30% moisture
#   - Agronomic optimal: 40–65% field capacity for most vegetables

mf_soil = {
    'dry':      fuzz.trapmf(u_soil, [0,  0,  15, 28]),   # must irrigate
    'low':      fuzz.trimf( u_soil, [20, 32, 45]),        # below optimal
    'adequate': fuzz.trimf( u_soil, [38, 52, 65]),        # target range
    'wet':      fuzz.trapmf(u_soil, [58, 72, 100, 100]),  # no irrigation needed
}

# ============================================================
#  MEMBERSHIP FUNCTIONS — VPD (input)
# ============================================================
# VPD = Vapor Pressure Deficit. Higher VPD → plants transpire more → need more water
# Breakpoints justified by:
#   - Predicted data: p5=7.7, p50=13.7, p95=24.8 hPa
#   - Ideal greenhouse VPD: 8–14 hPa (FAO guidelines)
#   - Above 20 hPa: severe plant stress

mf_vpd = {
    'low':       fuzz.trapmf(u_vpd, [0,  0,  6,  10]),   # low transpiration demand
    'optimal':   fuzz.trimf( u_vpd, [7,  12, 17]),        # ideal humidity balance
    'high':      fuzz.trimf( u_vpd, [13, 18, 25]),        # elevated stress
    'very_high': fuzz.trapmf(u_vpd, [21, 28, 50, 50]),   # severe water demand
}

# ============================================================
#  MEMBERSHIP FUNCTIONS — LIGHT (input)
# ============================================================
# Breakpoints justified by:
#   - Sensor data (nighttime): 3–6 raw units
#   - Predicted data: p25=26, p50=729, p95=1809 raw units
#   - Grow light needed when natural light < ~300 lux (daytime deficiency)

mf_light = {
    'dark':   fuzz.trapmf(u_light, [0,    0,   50,  200]),   # nighttime / overcast
    'dim':    fuzz.trimf( u_light, [100,  400,  800]),        # insufficient for growth
    'bright': fuzz.trapmf(u_light, [600, 1000, 2400, 2400]), # sufficient natural light
}

# ============================================================
#  MEMBERSHIP FUNCTIONS — FROST RISK (input)
# ============================================================
# Binary ML output (0 or 1) mapped to fuzzy sets
# Allows partial membership when the ML model outputs a probability

mf_frost = {
    'no_frost': fuzz.trapmf(u_frost, [0, 0,   0.3, 0.5]),
    'frost':    fuzz.trapmf(u_frost, [0.4, 0.6, 1.0, 1.0]),
}

# ============================================================
#  OUTPUT MEMBERSHIP FUNCTIONS
# ============================================================

# Heater intensity output (0–100)
mf_heater = {
    'off':    fuzz.trapmf(u_heater, [0, 0, 10, 25]),
    'low':    fuzz.trimf( u_heater, [15, 35, 55]),
    'medium': fuzz.trimf( u_heater, [45, 62, 78]),
    'high':   fuzz.trapmf(u_heater, [68, 82, 100, 100]),
}

# Pump duration output (0–120 seconds)
mf_pump = {
    'none':   fuzz.trapmf(u_pump, [0, 0, 5, 15]),
    'short':  fuzz.trimf( u_pump, [10, 30, 55]),
    'medium': fuzz.trimf( u_pump, [45, 65, 85]),
    'long':   fuzz.trapmf(u_pump, [75, 95, 120, 120]),
}

# Lamp power output (0–100)
mf_lamp = {
    'off':    fuzz.trapmf(u_lamp, [0, 0, 10, 25]),
    'dim':    fuzz.trimf( u_lamp, [15, 40, 60]),
    'full':   fuzz.trapmf(u_lamp, [50, 70, 100, 100]),
}

print('All membership functions defined successfully.')
```

## 4. Membership Function Visualization

Each plot shows the membership degree μ(x) ∈ [0, 1] for each linguistic term. Overlapping regions are intentional: a temperature of 20°C is simultaneously 30% COOL and 70% IDEAL — this is the core mechanism of fuzzy logic. The controller weights rules by these degrees, producing smooth output transitions.

```python
def plot_mfs(ax, universe, mf_dict, title, xlabel, color_map=None, show_data_line=None, vlines=None):
    """Reusable MF plotting function."""
    labels = list(mf_dict.keys())
    default_colors = list(COLORS.values())
    
    for i, (label, mf) in enumerate(mf_dict.items()):
        color = COLORS.get(label, default_colors[i % len(default_colors)]) if color_map is None else color_map[i]
        ax.plot(universe, mf, label=label.replace('_', ' ').title(), color=color, linewidth=2)
        ax.fill_between(universe, mf, alpha=0.08, color=color)
    
    if vlines:
        for v, vlabel in vlines:
            ax.axvline(v, color='gray', linewidth=1, linestyle='--', alpha=0.7)
            ax.text(v, 1.02, vlabel, fontsize=8, ha='center', color='gray')
    
    ax.set_title(title, fontsize=12, pad=8)
    ax.set_xlabel(xlabel)
    ax.set_ylabel('Membership degree μ(x)')
    ax.set_ylim(-0.05, 1.15)
    ax.legend(loc='upper right', fontsize=9)
    ax.grid(axis='y', alpha=0.3, linewidth=0.5)


# ── Figure 1: Input MFs ──────────────────────────────────────────────────────
fig, axes = plt.subplots(3, 2, figsize=(14, 13))
fig.suptitle('Figure 1: Mamdani FIS — Input Membership Functions', fontsize=14, y=1.01)

# Air temperature — show data mean from sensor readings
plot_mfs(axes[0,0], u_temp, mf_temp,
         'IN-1: Air Temperature',
         'Temperature (°C)',
         vlines=[(df_sensor['air_temp_c'].mean(), f'sensor\nmean={df_sensor["air_temp_c"].mean():.1f}°C')])

# Outdoor temperature — show Naryn context
plot_mfs(axes[0,1], u_out_temp, mf_out_temp,
         'IN-2: Outdoor Temperature (Naryn context)',
         'Temperature (°C)',
         vlines=[(df_pred['outdoor_temp_c'].mean(), f'mean={df_pred["outdoor_temp_c"].mean():.1f}°C'),
                 (df_pred['outdoor_temp_c'].quantile(0.05), f'p5={df_pred["outdoor_temp_c"].quantile(0.05):.1f}°C')])

# Soil moisture
plot_mfs(axes[1,0], u_soil, mf_soil,
         'IN-3: Soil Moisture (%)',
         'Soil Moisture (%)',
         vlines=[(df_pred['soil_moisture_pct'].mean(), f'mean={df_pred["soil_moisture_pct"].mean():.1f}%')])

# VPD
plot_mfs(axes[1,1], u_vpd, mf_vpd,
         'IN-4: Vapor Pressure Deficit (VPD)',
         'VPD (hPa)',
         vlines=[(df_pred['vpd_hpa'].mean(), f'mean={df_pred["vpd_hpa"].mean():.1f}')])

# Light
plot_mfs(axes[2,0], u_light, mf_light,
         'IN-5: Light Intensity (raw sensor units)',
         'Light (raw lux)')

# Frost risk
plot_mfs(axes[2,1], u_frost, mf_frost,
         'IN-6: Frost Risk (ML model output)',
         'Frost risk score')

plt.tight_layout()
plt.savefig('input_mfs.png', bbox_inches='tight', dpi=150)
plt.show()
print('Figure saved: input_mfs.png')
```

```python
# ── Figure 2: Output MFs ──────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
fig.suptitle('Figure 2: Mamdani FIS — Output Membership Functions', fontsize=14)

plot_mfs(axes[0], u_heater, mf_heater,
         'OUT-1: Heater Control Intensity',
         'Intensity (0–100)')

plot_mfs(axes[1], u_pump, mf_pump,
         'OUT-2: Pump Duration',
         'Duration (seconds)')

plot_mfs(axes[2], u_lamp, mf_lamp,
         'OUT-3: Lamp Power',
         'Power (0–100)')

plt.tight_layout()
plt.savefig('output_mfs.png', bbox_inches='tight', dpi=150)
plt.show()
print('Figure saved: output_mfs.png')
```

## 5. Fuzzy Rule Base

Rules are derived from two sources:
1. **Agronomic domain knowledge** — established greenhouse management guidelines
2. **Data patterns** — distributions observed in both datasets

Each rule has the form: `IF [antecedent conditions] THEN [consequent action]`

The antecedent uses AND (minimum operator) to combine conditions. All rules fire simultaneously; their outputs are aggregated by maximum before defuzzification.

```python
# ============================================================
#  FUZZY RULE BASE — HEATER CONTROL
# ============================================================
# Rules derived from:
#   - Optimal greenhouse temp: 18–26°C (most vegetable crops)
#   - Naryn outdoor min: -31.9°C (requires aggressive heating in winter)
#   - Frost damage begins below 5°C inside greenhouse

heater_rules = [
    # (antecedent_label, outdoor_label, frost_label, consequent_label, justification)
    ('cold',  'very_cold', 'frost',    'high',   'Emergency: frost risk + harsh outdoor + cold inside'),
    ('cold',  'very_cold', 'no_frost', 'high',   'Very cold indoor despite no frost signal — max heat'),
    ('cold',  'cold',      'frost',    'high',   'Frost risk present, cold outdoor, cold inside'),
    ('cold',  'cold',      'no_frost', 'medium', 'Cold inside, moderate outdoor cold'),
    ('cold',  'mild',      'no_frost', 'medium', 'Cold inside but mild outdoor — moderate heating'),
    ('cool',  'very_cold', 'frost',    'high',   'Frost risk from outside — pre-heat'),
    ('cool',  'cold',      'frost',    'high',   'Frost risk + cold outdoor, slightly cool inside'),
    ('cool',  'cold',      'no_frost', 'medium', 'Cool inside, cold outdoor — heat needed'),
    ('cool',  'mild',      'no_frost', 'low',    'Cool inside but mild outdoor — light heating'),
    ('ideal', 'very_cold', 'frost',    'medium', 'Frost risk from ML — preventive heating'),
    ('ideal', 'cold',      'no_frost', 'low',    'Temp OK but outdoor cold — maintain with low heat'),
    ('ideal', 'mild',      'no_frost', 'off',    'Temperature ideal, outdoor mild — no heating'),
    ('warm',  'mild',      'no_frost', 'off',    'Above ideal and outdoor mild — heater off'),
    ('warm',  'warm',      'no_frost', 'off',    'Warm inside and outside — heater off'),
    ('hot',   'mild',      'no_frost', 'off',    'Heat stress zone — heater definitely off'),
    ('hot',   'warm',      'no_frost', 'off',    'Heat stress — heater off, fan needed (separate rule)'),
]

print('=== HEATER RULE BASE ===')
print(f'Total rules: {len(heater_rules)}')
print()
print(f'{"#":<4} {"IF air_temp":<12} {"AND outdoor":<14} {"AND frost":<12} {"THEN heater":<13} {"Justification"}')
print('-' * 100)
for i, (t, o, f, out, just) in enumerate(heater_rules, 1):
    print(f'{i:<4} {t:<12} {o:<14} {f:<12} {out:<13} {just}')
```

```python
# ============================================================
#  FUZZY RULE BASE — PUMP CONTROL
# ============================================================
# Rules derived from:
#   - Capacitive sensor: raw 380–600 = dry, 277–380 = ideal, <270 = wet
#   - VPD high → more transpiration → more frequent watering
#   - High temperature increases evapotranspiration
#   - Dataset: sensor always shows ~484 raw (dry soil needs water)

pump_rules = [
    # (soil_label, vpd_label, temp_label, pump_output, justification)
    ('dry',      'very_high', 'hot',   'long',   'Severely dry + extreme VPD + hot = maximum irrigation'),
    ('dry',      'very_high', 'warm',  'long',   'Dry soil + very high VPD + warm — long watering'),
    ('dry',      'high',      'warm',  'long',   'Dry soil + high VPD + warm = extended irrigation'),
    ('dry',      'high',      'ideal', 'medium', 'Dry soil + elevated VPD — medium duration'),
    ('dry',      'optimal',   'ideal', 'medium', 'Dry soil at optimal VPD — moderate watering'),
    ('dry',      'optimal',   'cool',  'short',  'Dry soil but cool temp and optimal VPD — short burst'),
    ('dry',      'low',       'cool',  'short',  'Dry soil but low VPD — brief watering sufficient'),
    ('low',      'very_high', 'hot',   'medium', 'Low moisture + very high VPD + hot — medium irrigation'),
    ('low',      'high',      'warm',  'medium', 'Below optimal moisture + high VPD — water moderately'),
    ('low',      'optimal',   'ideal', 'short',  'Low moisture at normal conditions — short watering'),
    ('low',      'low',       'cool',  'none',   'Low moisture but low transpiration demand — wait'),
    ('adequate', 'very_high', 'hot',   'short',  'Adequate moisture but extreme demand — brief top-up'),
    ('adequate', 'high',      'warm',  'none',   'Adequate moisture — no irrigation needed'),
    ('adequate', 'optimal',   'ideal', 'none',   'Ideal conditions — no irrigation'),
    ('adequate', 'low',       'cool',  'none',   'Adequate moisture and low demand — no irrigation'),
    ('wet',      'very_high', 'hot',   'none',   'Soil wet — never irrigate regardless of VPD'),
    ('wet',      'optimal',   'ideal', 'none',   'Wet soil — no irrigation'),
    ('wet',      'low',       'cool',  'none',   'Wet soil — no irrigation'),
]

print('=== PUMP RULE BASE ===')
print(f'Total rules: {len(pump_rules)}')
print()
print(f'{"#":<4} {"IF soil":<12} {"AND VPD":<13} {"AND temp":<10} {"THEN pump":<12} {"Justification"}')
print('-' * 100)
for i, (s, v, t, out, just) in enumerate(pump_rules, 1):
    print(f'{i:<4} {s:<12} {v:<13} {t:<10} {out:<12} {just}')
```

## 6. Mamdani Inference Engine

The Mamdani inference process has four steps:
1. **Fuzzification** — compute membership degree of crisp inputs in each fuzzy set
2. **Rule evaluation** — apply AND (min) to combine antecedents; this is the rule firing strength
3. **Implication** — clip the output MF at the firing strength (Mamdani uses min)
4. **Aggregation** — combine all clipped output MFs using max
5. **Defuzzification** — compute centroid (center of gravity) of aggregated output

```python
def fuzzify(universe, mf_dict, crisp_value):
    """Step 1: Compute membership degree of a crisp value in all fuzzy sets."""
    return {label: fuzz.interp_membership(universe, mf, crisp_value)
            for label, mf in mf_dict.items()}


def mamdani_heater(air_temp, outdoor_temp, frost_risk_val):
    """
    Mamdani FIS for heater control.
    Returns: (crisp_output, aggregated_mf, activation_details)
    """
    # Step 1: Fuzzify
    mu_temp   = fuzzify(u_temp,     mf_temp,     air_temp)
    mu_out    = fuzzify(u_out_temp, mf_out_temp, outdoor_temp)
    mu_frost  = fuzzify(u_frost,    mf_frost,    frost_risk_val)

    # Step 2 & 3: Rule evaluation and implication
    activated = []
    for t_label, o_label, f_label, out_label, _ in heater_rules:
        # AND = minimum of all antecedent memberships
        strength = min(mu_temp[t_label], mu_out[o_label], mu_frost[f_label])
        if strength > 0:
            # Mamdani implication: clip output MF at firing strength
            clipped = np.fmin(strength, mf_heater[out_label])
            activated.append((out_label, strength, clipped))

    # Step 4: Aggregate (max of all clipped outputs)
    if not activated:
        return 0, np.zeros_like(u_heater), []
    agg = np.zeros_like(u_heater, dtype=float)
    for _, _, clipped in activated:
        agg = np.fmax(agg, clipped)

    # Step 5: Defuzzify (centroid method)
    if agg.sum() == 0:
        return 0, agg, activated
    crisp = fuzz.defuzz(u_heater, agg, 'centroid')
    return crisp, agg, activated


def mamdani_pump(soil_moisture, vpd, air_temp):
    """
    Mamdani FIS for pump duration control.
    Returns: (crisp_output_seconds, aggregated_mf, activation_details)
    """
    mu_soil = fuzzify(u_soil, mf_soil, soil_moisture)
    mu_vpd  = fuzzify(u_vpd,  mf_vpd,  vpd)
    mu_temp = fuzzify(u_temp, mf_temp, air_temp)

    activated = []
    for s_label, v_label, t_label, out_label, _ in pump_rules:
        strength = min(mu_soil[s_label], mu_vpd[v_label], mu_temp[t_label])
        if strength > 0:
            clipped = np.fmin(strength, mf_pump[out_label])
            activated.append((out_label, strength, clipped))

    if not activated:
        return 0, np.zeros_like(u_pump), []
    agg = np.zeros_like(u_pump, dtype=float)
    for _, _, clipped in activated:
        agg = np.fmax(agg, clipped)

    if agg.sum() == 0:
        return 0, agg, activated
    crisp = fuzz.defuzz(u_pump, agg, 'centroid')
    return crisp, agg, activated


def mamdani_lamp(light_val, air_temp):
    """
    Mamdani FIS for lamp control.
    Returns: (crisp_output, aggregated_mf, activation_details)
    """
    mu_light = fuzzify(u_light, mf_light, light_val)
    mu_temp  = fuzzify(u_temp,  mf_temp,  air_temp)

    activated = []
    for l_label, t_label, out_label, _ in lamp_rules:
        strength = min(mu_light[l_label], mu_temp[t_label])
        if strength > 0:
            clipped = np.fmin(strength, mf_lamp[out_label])
            activated.append((out_label, strength, clipped))

    if not activated:
        return 0, np.zeros_like(u_lamp), []
    agg = np.zeros_like(u_lamp, dtype=float)
    for _, _, clipped in activated:
        agg = np.fmax(agg, clipped)

    if agg.sum() == 0:
        return 0, agg, activated
    crisp = fuzz.defuzz(u_lamp, agg, 'centroid')
    return crisp, agg, activated


print('Inference functions defined.')
```

## 7. Defuzzification Visualization — Step-by-Step Examples

We demonstrate the complete Mamdani inference process for three representative scenarios drawn from real data conditions.

```python
def plot_inference_detail(ax_input_mfs, ax_output, universe_in, mf_dict_in, crisp_in,
                          universe_out, mf_dict_out, agg_out, crisp_out,
                          input_label, output_label, activated):
    """Visualise fuzzification + defuzzification for one input-output pair."""
    # Left panel: input with membership degrees shown
    for label, mf in mf_dict_in.items():
        color = COLORS.get(label, 'gray')
        ax_input_mfs.plot(universe_in, mf, color=color, linewidth=1.8, label=label)
        ax_input_mfs.fill_between(universe_in, mf, alpha=0.06, color=color)
    mu = fuzzify(universe_in, mf_dict_in, crisp_in)
    ax_input_mfs.axvline(crisp_in, color='black', linewidth=2, linestyle='-', label=f'input={crisp_in}')
    for label, degree in mu.items():
        if degree > 0.01:
            ax_input_mfs.plot([crisp_in], [degree], 'o', color=COLORS.get(label,'gray'), markersize=7)
    ax_input_mfs.set_title(input_label, fontsize=11)
    ax_input_mfs.set_ylabel('μ(x)')
    ax_input_mfs.set_ylim(-0.05, 1.15)
    ax_input_mfs.legend(fontsize=8, ncol=2)
    ax_input_mfs.grid(axis='y', alpha=0.3)

    # Right panel: aggregated output + crisp result
    for label, mf in mf_dict_out.items():
        color = COLORS.get(label, 'gray')
        ax_output.plot(universe_out, mf, color=color, linewidth=1.2, linestyle='--', alpha=0.4)
    if agg_out.sum() > 0:
        ax_output.fill_between(universe_out, agg_out, alpha=0.35, color='steelblue', label='Aggregated area')
        ax_output.plot(universe_out, agg_out, color='steelblue', linewidth=2)
        ax_output.axvline(crisp_out, color='red', linewidth=2.5, linestyle='-',
                          label=f'Centroid = {crisp_out:.1f}')
        ax_output.fill_betweenx([0, 1], 0, crisp_out, alpha=0.05, color='red')
    ax_output.set_title(output_label, fontsize=11)
    ax_output.set_ylabel('μ(x)')
    ax_output.set_ylim(-0.05, 1.15)
    ax_output.legend(fontsize=9)
    ax_output.grid(axis='y', alpha=0.3)


# ── Scenario definitions ─────────────────────────────────────────────────────
scenarios = [
    {
        'name': 'Scenario A — Frost Emergency (winter night in Naryn)',
        'air_temp':    10.0,    # cold greenhouse
        'outdoor_temp': -18.0,  # harsh Naryn winter (p5 of predicted data)
        'frost_risk':   1.0,    # ML model detected frost
        'soil_moisture': 12.0,  # dry soil
        'vpd':          20.0,   # high VPD
        'light':        5.0,    # nighttime
    },
    {
        'name': 'Scenario B — Normal Operations (spring day)',
        'air_temp':    23.0,    # sensor data mean (22.3°C)
        'outdoor_temp': 8.0,    # mild spring day
        'frost_risk':   0.0,
        'soil_moisture': 12.0,  # dry soil — actual sensor condition
        'vpd':          13.7,   # median VPD from predicted data
        'light':        4.0,    # sensor nighttime reading
    },
    {
        'name': 'Scenario C — Summer Heat Stress',
        'air_temp':    34.0,    # hot
        'outdoor_temp': 28.0,   # warm summer
        'frost_risk':   0.0,
        'soil_moisture': 8.0,   # very dry
        'vpd':          30.0,   # very high VPD (plant stress)
        'light':        1800.0, # bright summer day
    },
]

print('Scenarios defined. Running inference...')
for s in scenarios:
    h, _, _ = mamdani_heater(s['air_temp'], s['outdoor_temp'], s['frost_risk'])
    p, _, _ = mamdani_pump(s['soil_moisture'], s['vpd'], s['air_temp'])
    l, _, _ = mamdani_lamp(s['light'], s['air_temp'])
    s['heater_crisp'] = h
    s['pump_crisp'] = p
    s['lamp_crisp'] = l
    print(f"\n{s['name']}")
    print(f"  Heater: {h:.1f}/100  →  {'ON' if h > 40 else 'OFF'}")
    print(f"  Pump:   {p:.1f} sec  →  {'ON for ' + str(int(p)) + 's' if p > 5 else 'OFF'}")
    print(f"  Lamp:   {l:.1f}/100  →  {'ON' if l > 30 else 'OFF'}")
```

## 8. Control Surface Analysis

Control surfaces show how the FLC output varies as two inputs change simultaneously. This is a standard validation tool for fuzzy systems — a smooth, monotonic surface confirms that the rule base produces sensible and consistent behavior.

```python
# ── Figure 5: Control surfaces ───────────────────────────────────────────────
print('Computing control surfaces (this takes ~30 seconds)...')

# Surface 1: Heater — air_temp × outdoor_temp  (frost_risk=1, worst case)
temps_in  = np.arange(-5, 46, 2)
temps_out = np.arange(-35, 36, 2)
Z_heater  = np.zeros((len(temps_out), len(temps_in)))
for i, t_in in enumerate(temps_in):
    for j, t_out in enumerate(temps_out):
        v, _, _ = mamdani_heater(t_in, t_out, 0.5)  # medium frost risk
        Z_heater[j, i] = v

# Surface 2: Pump — soil_moisture × VPD  (air_temp=23, typical sensor reading)
soils = np.arange(0, 101, 4)
vpds  = np.arange(0, 51, 2)
Z_pump = np.zeros((len(vpds), len(soils)))
for i, s in enumerate(soils):
    for j, v in enumerate(vpds):
        out, _, _ = mamdani_pump(s, v, 23.0)
        Z_pump[j, i] = out

# Surface 3: Lamp — light × temp
lights = np.arange(0, 2401, 100)
temps3 = np.arange(-5, 46, 2)
Z_lamp = np.zeros((len(temps3), len(lights)))
for i, lg in enumerate(lights):
    for j, t in enumerate(temps3):
        out, _, _ = mamdani_lamp(lg, t)
        Z_lamp[j, i] = out

print('Done. Plotting...')

fig = plt.figure(figsize=(17, 5))
fig.suptitle('Figure 5: Mamdani FIS Control Surfaces', fontsize=13)

# Heater surface
ax1 = fig.add_subplot(131, projection='3d')
X1, Y1 = np.meshgrid(temps_in, temps_out)
surf1 = ax1.plot_surface(X1, Y1, Z_heater, cmap='coolwarm', edgecolor='none', alpha=0.9)
ax1.set_xlabel('Indoor temp (°C)', fontsize=9)
ax1.set_ylabel('Outdoor temp (°C)', fontsize=9)
ax1.set_zlabel('Heater (0–100)', fontsize=9)
ax1.set_title('Heater control surface\n(frost_risk=0.5)', fontsize=10)
fig.colorbar(surf1, ax=ax1, shrink=0.5)

# Pump surface
ax2 = fig.add_subplot(132, projection='3d')
X2, Y2 = np.meshgrid(soils, vpds)
surf2 = ax2.plot_surface(X2, Y2, Z_pump, cmap='YlOrBr', edgecolor='none', alpha=0.9)
ax2.set_xlabel('Soil moisture (%)', fontsize=9)
ax2.set_ylabel('VPD (hPa)', fontsize=9)
ax2.set_zlabel('Pump duration (s)', fontsize=9)
ax2.set_title('Pump control surface\n(air_temp=23°C)', fontsize=10)
fig.colorbar(surf2, ax=ax2, shrink=0.5)

# Lamp surface
ax3 = fig.add_subplot(133, projection='3d')
X3, Y3 = np.meshgrid(lights, temps3)
surf3 = ax3.plot_surface(X3, Y3, Z_lamp, cmap='plasma', edgecolor='none', alpha=0.9)
ax3.set_xlabel('Light (raw lux)', fontsize=9)
ax3.set_ylabel('Air temp (°C)', fontsize=9)
ax3.set_zlabel('Lamp power (0–100)', fontsize=9)
ax3.set_title('Lamp control surface', fontsize=10)
fig.colorbar(surf3, ax=ax3, shrink=0.5)

plt.tight_layout()
plt.savefig('control_surfaces.png', bbox_inches='tight', dpi=150)
plt.show()
print('Figure saved: control_surfaces.png')
```

## 9. Actuator Output Mapping (FLC → Arduino)

The Mamdani FLC produces crisp numeric outputs on [0, 100] or [0, 120]. The Arduino Mega only understands binary ON/OFF commands via MQTT. The Python automation engine acts as the translation layer:

| FLC Output | Meaning | Arduino Command |
|---|---|---|
| Heater > 40 | Heating required | `pump ON` (state=1) |
| Pump > 5 s | Irrigation required | `pump ON` for N seconds, then `OFF` |
| Lamp > 30 | Light supplement needed | `lamp ON` (state=1) |

```python
def flc_to_actuator_commands(heater_out, pump_out, lamp_out):
    """
    Translate FLC crisp outputs into binary Arduino MQTT commands.
    This is the bridge between the fuzzy controller and the hardware.
    """
    commands = {}

    # Heater: threshold at 40/100
    # Fan is complementary: above 70 (warm/hot zone), turn on fan instead
    if heater_out > 70:
        commands['heater'] = 1
        commands['fan']    = 0
        commands['note']   = 'Emergency heat'
    elif heater_out > 40:
        commands['heater'] = 1
        commands['fan']    = 0
        commands['note']   = 'Normal heating'
    elif heater_out < 15:
        # Low heater output AND if indoor was warm/hot → fan on
        commands['heater'] = 0
        commands['fan']    = 0
        commands['note']   = 'Heater off'
    else:
        commands['heater'] = 0
        commands['fan']    = 0
        commands['note']   = 'Idle'

    # Pump: duration in seconds (maps directly to sleep timer in automation engine)
    commands['pump_on']       = pump_out > 5
    commands['pump_duration'] = int(pump_out) if pump_out > 5 else 0

    # Lamp: threshold at 30/100
    commands['lamp'] = 1 if lamp_out > 30 else 0

    return commands


# ── Test all scenarios ───────────────────────────────────────────────────────
print('=' * 70)
print('FLC OUTPUT → ACTUATOR COMMAND TRANSLATION')
print('=' * 70)

for s in scenarios:
    h, _, _ = mamdani_heater(s['air_temp'], s['outdoor_temp'], s['frost_risk'])
    p, _, _ = mamdani_pump(s['soil_moisture'], s['vpd'], s['air_temp'])
    l, _, _ = mamdani_lamp(s['light'], s['air_temp'])
    cmds = flc_to_actuator_commands(h, p, l)

    print(f"\n{s['name']}")
    print(f"  Inputs:  air_temp={s['air_temp']}°C, outdoor={s['outdoor_temp']}°C, "
          f"frost={s['frost_risk']}, soil={s['soil_moisture']}%, "
          f"vpd={s['vpd']}, light={s['light']}")
    print(f"  FLC out: heater={h:.1f}, pump={p:.1f}s, lamp={l:.1f}")
    print(f"  MQTT:    heater→{cmds['heater']} ({cmds['note']}), "
          f"fan→{cmds['fan']}, "
          f"pump→{'ON for '+str(cmds['pump_duration'])+'s' if cmds['pump_on'] else 'OFF'}, "
          f"lamp→{cmds['lamp']}")
print()
```

## 10. System Evaluation on Real Sensor Data

We apply the FLC to the actual sensor readings from Aruuke's dataset and evaluate the controller's output distribution over time.

```python
# Apply FLC to the real sensor dataset
# For inputs not in sensor data (outdoor_temp, frost_risk, vpd),
# we use the median values from Alfiia's predicted dataset.
# In the actual pipeline, these come from the ML prediction module.

outdoor_temp_median = df_pred['outdoor_temp_c'].median()   # 1.96°C (Naryn)
frost_risk_median   = df_pred['frost_risk'].median()       # 1.0 (majority frost)
vpd_median          = df_pred['vpd_hpa'].median()          # 13.7 hPa

results = []
for _, row in df_sensor.iterrows():
    h, _, _ = mamdani_heater(row['air_temp_c'], outdoor_temp_median, frost_risk_median)
    p, _, _ = mamdani_pump(row['soil_moisture_pct'] if pd.notna(row['soil_moisture_pct'])
                           else 5.0,   # fallback: use low moisture if null
                           vpd_median, row['air_temp_c'])
    l, _, _ = mamdani_lamp(row['light_raw'], row['air_temp_c'])
    cmds = flc_to_actuator_commands(h, p, l)
    results.append({
        'timestamp':      row['timestamp'],
        'air_temp_c':     row['air_temp_c'],
        'soil_moisture':  row.get('soil_moisture_pct', 5.0),
        'light_raw':      row['light_raw'],
        'heater_flc':     h,
        'pump_flc':       p,
        'lamp_flc':       l,
        'heater_cmd':     cmds['heater'],
        'pump_duration':  cmds['pump_duration'],
        'lamp_cmd':       cmds['lamp'],
    })

df_results = pd.DataFrame(results)

print('FLC applied to full sensor dataset.')
print(f'Records processed: {len(df_results)}')
print(f"\nActuator activation summary (over {len(df_results)} readings):")
print(f"  Heater ON:  {df_results['heater_cmd'].sum():>4} times  ({df_results['heater_cmd'].mean()*100:.1f}% of readings)")
print(f"  Pump ON:    {(df_results['pump_duration']>0).sum():>4} times  ({(df_results['pump_duration']>0).mean()*100:.1f}% of readings)")
print(f"  Pump mean duration: {df_results[df_results['pump_duration']>0]['pump_duration'].mean():.1f}s  "
      f"(when activated)")
print(f"  Lamp ON:    {df_results['lamp_cmd'].sum():>4} times  ({df_results['lamp_cmd'].mean()*100:.1f}% of readings)")
```

## 11. Sensitivity Analysis

We test how sensitive the FLC outputs are to changes in each input variable. A well-designed fuzzy system should show smooth, monotonic sensitivity — not abrupt jumps.

```python
# ── Figure 8: Sensitivity analysis ───────────────────────────────────────────
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
fig.suptitle('Figure 8: FLC Sensitivity Analysis — Output vs Each Input Variable', fontsize=13)

# Heater vs indoor temperature (outdoor=-5, frost=0.5)
temps_sweep = np.arange(-5, 46, 0.5)
heater_vs_temp = [mamdani_heater(t, -5, 0.5)[0] for t in temps_sweep]
axes[0,0].plot(temps_sweep, heater_vs_temp, color='#E24B4A', linewidth=2)
axes[0,0].axhline(40, color='gray', linestyle='--', linewidth=1, label='ON threshold')
axes[0,0].set_xlabel('Indoor temperature (°C)')
axes[0,0].set_ylabel('Heater output (0–100)')
axes[0,0].set_title('Heater vs Indoor Temp\n(outdoor=−5°C, frost_risk=0.5)')
axes[0,0].legend(fontsize=9)
axes[0,0].grid(alpha=0.3)

# Heater vs frost risk (indoor=15, outdoor=-10)
frost_sweep = np.arange(0, 1.05, 0.05)
heater_vs_frost = [mamdani_heater(15, -10, f)[0] for f in frost_sweep]
axes[0,1].plot(frost_sweep, heater_vs_frost, color='#E24B4A', linewidth=2)
axes[0,1].axhline(40, color='gray', linestyle='--', linewidth=1, label='ON threshold')
axes[0,1].set_xlabel('Frost risk (0=none, 1=frost)')
axes[0,1].set_ylabel('Heater output (0–100)')
axes[0,1].set_title('Heater vs Frost Risk\n(indoor=15°C, outdoor=−10°C)')
axes[0,1].legend(fontsize=9)
axes[0,1].grid(alpha=0.3)

# Pump vs soil moisture (vpd=13, temp=23)
soil_sweep = np.arange(0, 101, 1)
pump_vs_soil = [mamdani_pump(s, 13, 23)[0] for s in soil_sweep]
axes[0,2].plot(soil_sweep, pump_vs_soil, color='#3B8BD4', linewidth=2)
axes[0,2].axhline(5, color='gray', linestyle='--', linewidth=1, label='ON threshold')
axes[0,2].set_xlabel('Soil moisture (%)')
axes[0,2].set_ylabel('Pump duration (s)')
axes[0,2].set_title('Pump duration vs Soil Moisture\n(VPD=13 hPa, temp=23°C)')
axes[0,2].legend(fontsize=9)
axes[0,2].grid(alpha=0.3)

# Pump vs VPD (soil=15, temp=23)
vpd_sweep = np.arange(0, 51, 0.5)
pump_vs_vpd = [mamdani_pump(15, v, 23)[0] for v in vpd_sweep]
axes[1,0].plot(vpd_sweep, pump_vs_vpd, color='#3B8BD4', linewidth=2)
axes[1,0].axhline(5, color='gray', linestyle='--', linewidth=1, label='ON threshold')
axes[1,0].set_xlabel('VPD (hPa)')
axes[1,0].set_ylabel('Pump duration (s)')
axes[1,0].set_title('Pump duration vs VPD\n(soil=15%, temp=23°C)')
axes[1,0].legend(fontsize=9)
axes[1,0].grid(alpha=0.3)

# Lamp vs light input (temp=22)
light_sweep = np.arange(0, 2401, 20)
lamp_vs_light = [mamdani_lamp(lg, 22)[0] for lg in light_sweep]
axes[1,1].plot(light_sweep, lamp_vs_light, color='#EF9F27', linewidth=2)
axes[1,1].axhline(30, color='gray', linestyle='--', linewidth=1, label='ON threshold')
axes[1,1].set_xlabel('Light intensity (raw lux)')
axes[1,1].set_ylabel('Lamp output (0–100)')
axes[1,1].set_title('Lamp output vs Light Sensor\n(temp=22°C)')
axes[1,1].legend(fontsize=9)
axes[1,1].grid(alpha=0.3)

# Lamp vs temperature (light=5, dark scenario)
temp_sweep2 = np.arange(-5, 46, 0.5)
lamp_vs_temp = [mamdani_lamp(5, t)[0] for t in temp_sweep2]
axes[1,2].plot(temp_sweep2, lamp_vs_temp, color='#EF9F27', linewidth=2)
axes[1,2].axhline(30, color='gray', linestyle='--', linewidth=1, label='ON threshold')
axes[1,2].set_xlabel('Air temperature (°C)')
axes[1,2].set_ylabel('Lamp output (0–100)')
axes[1,2].set_title('Lamp output vs Temperature\n(light=5 raw, nighttime)')
axes[1,2].legend(fontsize=9)
axes[1,2].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('sensitivity_analysis.png', bbox_inches='tight', dpi=150)
plt.show()
print('Figure saved: sensitivity_analysis.png')
```

## 12. Summary and Methodology Justification

This section consolidates the justification for the Mamdani FIS as the decision-making module in the greenhouse control pipeline.

```python
print('=' * 70)
print('MAMDANI FIS — COMPLETE SYSTEM SUMMARY')
print('=' * 70)

print("""
SYSTEM ARCHITECTURE
-------------------
The FIS sits between two upstream modules and the physical actuators:

  [Kalman-smoothed sensor data]     ──┐
  [ML predictions + frost_risk]     ──┤→ Mamdani FIS → [Python bridge] → MQTT → Arduino
  [Schedule / context]              ──┘

INPUTS (6 variables)
--------------------
  IN-1  air_temp_c         Range: −5 to 45°C     5 linguistic terms
  IN-2  outdoor_temp_c     Range: −35 to 35°C    4 linguistic terms  (Naryn climate)
  IN-3  soil_moisture_pct  Range: 0 to 100%      4 linguistic terms
  IN-4  vpd_hpa            Range: 0 to 50 hPa    4 linguistic terms
  IN-5  light_raw          Range: 0 to 2400 raw  3 linguistic terms
  IN-6  frost_risk         Range: 0 to 1         2 linguistic terms  (ML output)

OUTPUTS (3 controllers)
-----------------------
  OUT-1  heater_intensity   Range: 0–100   4 terms  → threshold >40 → MQTT ON
  OUT-2  pump_duration      Range: 0–120s  4 terms  → duration > 5s → MQTT ON+sleep
  OUT-3  lamp_power         Range: 0–100   3 terms  → threshold >30 → MQTT ON
""")

print(f'RULE BASE')
print(f'---------')
print(f'  Heater controller: {len(heater_rules)} rules')
print(f'  Pump controller:   {len(pump_rules)} rules')
print(f'  Lamp controller:   {len(lamp_rules)} rules')
print(f'  Total:             {len(heater_rules)+len(pump_rules)+len(lamp_rules)} rules')

print("""
INFERENCE METHOD
----------------
  Type:            Mamdani (max-min)
  AND operator:    Minimum (conservative, avoids over-activation)
  Implication:     Minimum (clips output MF at rule firing strength)
  Aggregation:     Maximum (combines all active rule outputs)
  Defuzzification: Centroid (center of gravity — smooth, balanced output)

MF SHAPE CHOICES
----------------
  Interior terms:  Triangular — single clear peak, symmetric overlap
  Boundary terms:  Trapezoidal — flat top ensures saturation at extremes
                   (e.g., below −5°C is always fully COLD)

BREAKPOINT JUSTIFICATION
------------------------
  All MF breakpoints are grounded in:
  (a) Statistical analysis of sensor_data_march30.csv (504 real readings)
  (b) Statistical analysis of predicted_data_march30.csv (35,545 samples)
  (c) Naryn, Kyrgyzstan climate context (outdoor temp range −31.9 to 34.4°C)
  (d) FAO greenhouse management guidelines (optimal temp 18–26°C,
      VPD 8–14 hPa, soil moisture 40–65% field capacity)
  (e) Sensor calibration: capacitive soil sensor raw 380–600 = dry,
      277–380 = ideal (from automation_engine.py calibration comments)

KEY DESIGN ADVANTAGES OVER HARD-THRESHOLD SYSTEM
------------------------------------------------
  1. Smooth transitions: no abrupt ON/OFF switching at a single threshold
  2. Multi-input reasoning: pump duration adapts to VPD, not just soil alone
  3. Frost emergency: frost_risk from ML model directly triggers
     proportional heater response regardless of indoor temperature
  4. Naryn-specific: outdoor temperature input makes heating effort
     proportional to the actual thermal load — critical for −30°C winters
  5. Interpretable: every decision can be traced to a readable IF-THEN rule
""")
```
